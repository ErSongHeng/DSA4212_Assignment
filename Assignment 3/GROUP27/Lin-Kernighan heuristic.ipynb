```python
import numpy as np
import elkai
import math
import time
```

```python
cities = np.load("cities.npy", allow_pickle = True)

dist_matrix = [[math.dist(c1, c2) for c2 in cities] for c1 in cities]
```

```python
t0=time.time()
# Get the optimal route
solution = elkai.solve_float_matrix(dist_matrix, runs=10)

# Calculate the distance
total_dist = 0
for i in range(0, len(solution)-1):
    start = solution[i]
    end = solution[i+1]
    total_dist+=dist_matrix[start][end]
    
    if i == len(solution):
        start = solution[i+1]
        end = solution[0]
        total_dist+=dist_matrix[start][end]
        
t1=time.time()
        
print("Path is ", solution)
print("Distance is ", total_dist)
print("Total time taken: ", t1-t0, " seconds")
```

```python
import numpy as np

# Calculate the cost change
def cost_change(cost_mat, n1, n2, n3, n4):
    return cost_mat[n1][n3] + cost_mat[n2][n4] - cost_mat[n1][n2] - cost_mat[n3][n4]

# 2-opt algorithm to improve upon the final route
def two_opt(route, cost_mat):
    best = route
    improved = True
    while improved:
        improved = False
        for i in range(1, len(route) - 2):
            for j in range(i + 1, len(route)):
                if j - i == 1: continue
                if cost_change(cost_mat, best[i - 1], best[i], best[j - 1], best[j]) < 0:
                    best[i:j] = best[j - 1:i - 1:-1]
                    improved = True
        route = best
    return best

best_route = two_opt(solution, dist_matrix)
```

```python
# Calculate the distance

total_dist = 0
for i in range(0, len(best_route)-1):
    start = best_route[i]
    end = best_route[i+1]
    total_dist+=dist_matrix[start][end]
    
    if i == len(best_route):
        start = best_route[i+1]
        end = best_route[0]
        total_dist+=dist_matrix[start][end]
        
print(best_route)
print(total_dist)
```
